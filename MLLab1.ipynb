```python
ages = [19, 22, 19, 24, 20, 25, 26, 24, 25, 24]

ages.sort()
min_age = ages[0]
max_age = ages[-1]

ages.extend([min_age, max_age])

n = len(ages)
if n % 2 == 0:
    median_age = (ages[n//2 - 1] + ages[n//2]) / 2
else:
    median_age = ages[n//2]

average_age = sum(ages) / len(ages)

age_range = max_age - min_age

print("Sorted Ages:", ages)
print("Minimum Age:", min_age)
print("Maximum Age:", max_age)
print("Median Age:", median_age)
print("Average Age:", average_age)
print("Range of Ages:", age_range)
```

```python
import math

# Given radius
radius = 30

# Calculate the area of the circle
area_of_circle = math.pi * radius ** 2

# Calculate the circumference of the circle
circum_of_circle = 2 * math.pi * radius

print("Area of the circle:", area_of_circle)
print("Circumference of the circle:", circum_of_circle)

# Take radius as user input
user_radius = float(input("Enter the radius of the circle (in meters): "))

# Calculate the area using user input radius
area_of_circle_user_input = math.pi * user_radius ** 2

print("Area of the circle with radius", user_radius, "is:", area_of_circle_user_input)
```

```python
# Read the number of students from the user
N = int(input("Enter the number of students: "))

# Initialize an empty list to store weights in pounds
weights_lbs = []

# Read weights of N students from the user
for i in range(N):
    weight_lbs = float(input(f"Enter weight of student {i+1} (in lbs): "))
    weights_lbs.append(weight_lbs)

# Convert weights from pounds to kilograms and store them in a separate list
weights_kg = []
for weight_lbs in weights_lbs:
    weight_kg = weight_lbs * 0.453592  # Conversion factor: 1 lb = 0.453592 kg
    weights_kg.append(round(weight_kg, 2))  # Round to 2 decimal places

# Print the converted weights
print("Weights in kilograms:", weights_kg)
```
